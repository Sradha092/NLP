```python
complete_paragraph = ' '.join(product_reviews_2.words('Canon_PowerShot_SD500.txt'))
```

```python
complete_paragraph
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize
```

```python
sent_tokenize(complete_paragraph)
```

```python
for sent in sent_tokenize(complete_paragraph):
    print(sent)
```

```python
word_tokenize(complete_paragraph)
```

```python
for word in word_tokenize(complete_paragraph):
    print(word)
```

```python
white_token = nltk.tokenize.WhitespaceTokenizer()
```

```python
white_token.tokenize(complete_paragraph)
```

```python
for white in white_token.tokenize(complete_paragraph):
    print(white)
```

## Word Punctuation

```python
word_punc = nltk.tokenize.WordPunctTokenizer()
```

```python
my_word = word_punc.tokenize(complete_paragraph)
my_word
```

```python
from nltk.corpus import stopwords
```

```python
stop_words = stopwords.words('english')
stop_words
```

```python
sent_filter = [w for w in my_word if not w in stop_words]
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
word_stop = nltk.corpus.stopwords.words('english')
```

```python
word_stop
```

```python
wcloud = WordCloud(max_words=100, stopwords=word_stop)
wcloud.generate(complete_paragraph)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,14))
plt.imshow(wcloud)
plt.show()
```

```python
wcloud = WordCloud(max_words=100, stopwords=word_stop+['.','camera','picture','SD500','one'])
wcloud.generate(complete_paragraph)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,14))
plt.imshow(wcloud)
plt.show()
```

```python
wcloud = WordCloud(max_words=100, stopwords=word_stop, background_color='white')
wcloud.generate(complete_paragraph)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,14))
plt.imshow(wcloud)
plt.show()
```
